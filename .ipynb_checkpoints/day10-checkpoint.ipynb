```python
from z3 import *

# set the logic
set_option('model', True)
set_option('smt.arith.nl', True)

buttons = [[3],[1,3],[2],[2,3],[0,2],[0,1]]
target = [3,5,4,7]

def solveProblem (buttons,target):
    matrix = [[0 for i in range(11)] for j in target]
    for i,t in enumerate(target):
        matrix[i][10] = t
    for i,b in enumerate(buttons):
        for s in b:
            matrix[s][i] = 1
    X = [ Int('x%s' % i) for i in range(11) ]
    GT0 = [X[i]>=0 for i in range(11)]
    EQ = [C[0]*X[0]+C[1]*X[1]+C[2]*X[2]+C[3]*X[3]+C[4]*X[4]+C[5]*X[5]+C[6]*X[6]
          +C[7]*X[7]+C[8]*X[8]+C[9]*X[9] == C[10] for C in matrix ]
    s = Optimize()
    # add constraints
    for e in EQ:
        s.add(e)
    for e in GT0:
        s.add(e)
    s.minimize(X[0]+X[1]+X[2]+X[3]+X[4]+X[5]+X[6]+X[7]+X[8]+X[9])
    s.check()
    m = s.model()
    summary = 0
    for d in m.decls():
        #print ("%s = %s" % (d.name(), m[d]))
        summary += m[d].as_long()
    return summary

solveProblem(buttons,target)
```

```python
import re
filename="data/d10-test.txt"
#re1 = r'^\\[(.+)\\] (\\(.+\\)\\s)\\{(.+)\\}$'
re1 = r'^\[(.+)\] (\(.+\)\s)\{(.+)\}$'
lines =[line for line in open(filename).read().strip().split('\n')]
for line in lines:
    res = re.match(re1,line)
    buttons = [list(map (lambda s: int(s), b[1:-1].split(','))) for b in res.groups()[1].strip().split(' ')]
    print(buttons)
    target = res.groups()[2].split(',')
    r = solveProblem(buttons,target)
    print(r)
```
